```python
# Generate a list of squares
squares = [i**2 for i in range(1, 11)]
squares
```

```python
# Sample sales data analysis
sales_data = [
    {"product": "Laptop", "price": 999, "quantity": 15},
    {"product": "Mouse", "price": 25, "quantity": 50},
    {"product": "Keyboard", "price": 75, "quantity": 30},
    {"product": "Monitor", "price": 299, "quantity": 20},
    {"product": "Headphones", "price": 149, "quantity": 25}
]

print("Sample Data Analysis")
print("=" * 20)
print("Dataset: Sales Data")
print(f"Records: {len(sales_data)}")
print("Columns: Product, Price, Quantity")
print()
print(f"{'Product':<12} | {'Price':<5} | {'Quantity':<8}")
print("-" * 12 + "|" + "-" * 7 + "|" + "-" * 10)

total_revenue = 0
for item in sales_data:
    revenue = item["price"] * item["quantity"]
    total_revenue += revenue
    print(f"{item['product']:<12} | ${item['price']:<4} | {item['quantity']:<8}")

print(f"\nTotal Revenue: ${total_revenue:,}")
```

```python
# Calculate statistics for the squares
def calculate_stats(numbers):
    return {
        "mean": sum(numbers) / len(numbers),
        "median": sorted(numbers)[len(numbers) // 2],
        "sum": sum(numbers),
        "min": min(numbers),
        "max": max(numbers),
        "count": len(numbers)
    }

stats = calculate_stats(squares)
stats
```
