# Data Analysis Example

This notebook demonstrates basic data analysis operations.

```python
import pandas as pd
import numpy as np

# Create sample data
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'age': [25, 30, 35, 28],
    'score': [85, 92, 78, 96]
}
```

```python
# Basic statistics
ages = data['age']
scores = data['score']

print(f"Average age: {np.mean(ages)}")
print(f"Maximum score: {max(scores)}")
print(f"Total participants: {len(data['name'])}")
```

```python
# Create and display DataFrame
df = pd.DataFrame(data)
df
```

```python
# Generate grade list
grades = []
for name, score in zip(data['name'], data['score']):
    grade = 'A' if score >= 90 else 'B' if score >= 80 else 'C'
    grades.append(f"{name}: Grade {grade}")

grades
```
